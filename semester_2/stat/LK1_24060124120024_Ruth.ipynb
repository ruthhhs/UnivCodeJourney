## Properti 1: Menambah atau Mengurangi semua Dataset dengan Nilai Konstanta

### Langkah 1: Memodifikasi dataset

Misal kita lakukan modifikasi kepada setiap data di dataset dengan **penjumlahan 2** sebagai modifikasi1 dan **pengurangan 3** sebagai modifikasi2.

```python
# Menambahkan 2 ke setiap elemen sebagai modifikasi1
df_modif1 = df+2
# Menampilkan 5 baris pertama dari dataset modifikasi1
print("Dataset modifikasi1:")
print(df_modif1.head())

# Mengurangkan 3 ke setiap elemen sebagai modifikasi2
df_modif2 = df-3
# Menampilkan 5 baris pertama dari dataset modifikasi2
print("\nDataset modifikasi2:")
print(df_modif2.head())
```

### Langkah 2: Menghitung nilai mean dan median dari data yang sudah dimodifikasi

```python
# Menghitung mean dan median untuk modifikasi1 di dataset "Tipe 1"
mean_modif1 = df_modif1['Tipe1'].mean()
median_modif1 = df_modif1['Tipe1'].median()
# Menampilkan hasil mean dan median
print("Mean dari modifikasi 1 dataset Tipe 1:", mean_modif1)
print("Median dari modifikasi 1 dataset Tipe 1:", median_modif1)

# Menghitung mean dan median untuk modifikasi2 di dataset "Tipe 1"
mean_modif2 = df_modif2['Tipe1'].mean()
median_modif2 = df_modif2['Tipe1'].median()
# Menampilkan hasil mean dan median
print("\nMean dari modifikasi 2 dataset Tipe 1:", mean_modif2)
print("Median dari modifikasi 2 dataset Tipe 1:", median_modif2)
```

### Langkah 3: Analisis

Dari beberapa perhitungan diatas kita peroleh hasil sebagai berikut:

Properti asli                : mean 7.83 dan median 7

Modifikasi 1 (penjumlahan 2) : mean 9.83 dan median 9 --> ($9.83 = 7.83 + 2$ dan $9 = 7 + 2$)

Modifikasi 2 (pengurangan 3) : mean 4.83 dan median 4 --> ($4.83 = 7.83 - 3$ dan $4 = 7 - 3$)

Jika melihat polanya, bisa kita simpulkan bahwa mean dan median dari penjumlahan (atau pengurangan) konstanta kepada setiap data dalam dataset sama dengan mean dan median dari dataset ditambah (atau dikurang) konstanta tersebut.

Atau bisa dituliskan secara matematis sebagai berikut:

$$
\frac{\sum x_i \pm c}{n} = \frac{\sum x_i}{n} \pm c
$$

$$
\bar{x}_{\text{baru}} = \bar{x} \pm c
$$

Pada median, sebagai berikut:

$$
\text{Me}_{\text{baru}} = \text{Me} \pm c
$$




## Properti 2: Mengalikan atau Membagi semua Dataset dengan Nilai Konstanta

### Langkah 1: Memodifikasi dataset

Misal kita lakukan modifikasi kepada setiap data di dataset dengan **perkalian 2** sebagai modifikasi1 dan **pembagian 3** sebagai modifikasi2.

```python
# Menglikan 2 ke setiap elemen sebagai modifikasi1
df_modif1 = df*2
# Menampilkan 5 baris pertama dari dataset modifikasi1
print("Dataset modifikasi1:")
print(df_modif1.head())

# Membagi 3 ke setiap elemen sebagai modifikasi2
df_modif2 = df/3
# Menampilkan 5 baris pertama dari dataset modifikasi2
print("\nDataset modifikasi2:")
print(df_modif2.head())
```

### Langkah 2: Menghitung nilai mean dan median dari data yang sudah dimodifikasi

```python
# Menghitung mean dan median untuk modifikasi1 di dataset "Tipe 1"
mean_modif1 = df_modif1['Tipe1'].mean()
median_modif1 = df_modif1['Tipe1'].median()
# Menampilkan hasil mean dan median
print("Mean dari modifikasi 1 dataset Tipe 1:", mean_modif1)
print("Median dari modifikasi 1 dataset Tipe 1:", median_modif1)

# Menghitung mean dan median untuk modifikasi2 di dataset "Tipe 1"
mean_modif2 = df_modif2['Tipe1'].mean()
median_modif2 = df_modif2['Tipe1'].median()
# Menampilkan hasil mean dan median
print("\nMean dari modifikasi 2 dataset Tipe 1:", mean_modif2)
print("Median dari modifikasi 2 dataset Tipe 1:", median_modif2)

```

### Langkah 3: Analisis

Dari beberapa perhitungan diatas kita peroleh hasil sebagai berikut:

Properti asli                : mean 7.83 dan median 7

Modifikasi 1 (perkalian 2) : mean 15.67 dan median 14 --> ($15.67 = 7.83\times2$ dan $14 = 7\times2$)

Modifikasi 2 (pembagian 3) : mean 2.61 dan median 2.33 --> ($2.61 = 7.83/3$ dan $2.33 = 7/3$)

Jika melihat polanya, bisa kita simpulkan bahwa mean dan median dari perkalian (atau pembagian) konstanta kepada setiap data dalam dataset sama dengan mean dan median dari dataset dikali (atau dibagi) konstanta tersebut.

Atau bisa dituliskan secara matematis sebagai berikut:

$$
\frac{\sum x_i \times c}{n} = \frac{\sum x_i}{n} \times c
$$

$$
\bar{x}_{\text{baru}} = \bar{x} \times c
$$

Pada median, sebagai berikut:

$$
\text{Me}_{\text{baru}} = \text{Me} \times c
$$



## Properti 3: Sensitif terhadap Outlier

### Langkah 1: Memodifikasi dataset

Misal kita lakukan modifikasi dimana terdapat mahasiswa baru super jenius yang mendapat nilai 1000 di tiap tipe ujian, kita tambahkan **1 baris data dengan nilai 1000** ke dalam dataset.

```python
# Menambahkan satu baris data dengan nilai 1000 untuk setiap kolom tipe ujian
new_row = pd.DataFrame({col: [1000] for col in df.columns})
df = pd.concat([df, new_row], ignore_index=True)

# Menghapus baris terakhir yang ditambahkan
# Gunakan agar tidak ada baris baru ganda jika program dijalankan berulang kali
# df = df.iloc[:-1] 

# Menampilkan 5 baris terakhir dari dataset modifikasi outlier
print("\nDataset modifikasi outlier:")
print(df.tail())
```

### Langkah 2: Menghitung nilai mean dan median dari data yang sudah dimodifikasi

```python
# Menghitung mean dan median untuk modifikasiol di dataset "Tipe 1"
mean_modifol = df['Tipe1'].mean()
median_modifol = df['Tipe1'].median()
# Menampilkan hasil mean dan median
print("Mean dari modifikasi outlier dataset Tipe 1:", mean_modifol)
print("Median dari modifikasi outlier dataset Tipe 1:", median_modifol)
```

### Langkah 3: Analisis

Dari beberapa perhitungan diatas kita peroleh hasil sebagai berikut:

Properti asli      : mean 7.83 dan median 7

Modifikasi outlier : mean 25.24 dan median 7

Dari data tersebut bisa kita lihat bahwa mean berubah sangat drastis, sedangkan median tidak berubah. Ini karena mean menjumlahkan seluruh data dan membaginya dengan jumlah data. Penambahan outlier akan menghasilkan nilai penjumlahan seluruh data yang sangat besar, sedangkan pembaginya hanya bertambah satu, sehingga hasil pembagian akan sangat besar. Sedangkan median hanya mengambil data di tengah dataset, sehingga outlier bisa dihiraukan.

Dari sini bisa kita simpulkan bahwa outlier sangat mempengaruhi mean, sedangkan untuk median tidak.
